# Color-Blind Friendly Color Guide

## Understanding Color Blindness
- **8-10% of men** and **0.5% of women** have some form of color vision deficiency
- **Most common types:**
  - **Deuteranopia** (green-blind) - most common
  - **Protanopia** (red-blind) 
  - **Tritanopia** (blue-blind) - rare

## Best Color-Blind Friendly Palettes

### 1. **Recommended Scientific Palette** (What we'll use):
- **Blue**: `#0173B2` - Safe for all types
- **Orange**: `#DE8F05` - High contrast alternative to red
- **Green**: `#029E73` - Distinguishable green
- **Red**: `#CC78BC` - Pink-red, safer than pure red
- **Gray**: `#56B4E9` - Light blue-gray
- **Dark**: `#2E3440` - Professional dark

### 2. **Paul Tol's Palette** (Highly recommended):
- Specifically designed by a color-blind scientist
- Colors: `#4477AA`, `#EE6677`, `#228833`, `#CCBB44`, `#66CCEE`, `#AA3377`

### 3. **Viridis-inspired** (Perceptually uniform):
- Good for continuous data
- Colors transition smoothly for all vision types

## Design Principles:
1. **Use different shapes/patterns** in addition to colors
2. **Avoid red-green combinations** 
3. **Test with grayscale** - should still be distinguishable
4. **High contrast** between adjacent elements
5. **Use colorbrewer2.org** for palette testing

# Alternative Color-Blind Friendly Colors
*Avoiding blue, green, purple, red, black, yellow*

## Excellent Options:

### **Orange Family** (Highly Recommended):
- **Bright Orange**: `#FF8C00` - Very distinctive and safe
- **Dark Orange**: `#FF7F00` - Professional look
- **Burnt Orange**: `#CC5500` - Sophisticated

### **Brown/Tan Family**:

- **Saddle Brown**: `#8B4513` - Classic scientific color
- **Sandy Brown**: `#F4A460` - Light but visible

### **Gray Family** (Neutral but effective):
- **Dark Gray**: `#696969` - Professional
- **Slate Gray**: `#708090` - Modern look
- **Dim Gray**: `#666666` - Good contrast

### **Teal/Cyan Family** (Blue-green but distinguishable):
- **Dark Turquoise**: `#00CED1` - Distinctive
- **Light Sea Green**: `#20B2AA` - Safe alternative
- **Medium Turquoise**: `#48D1CC` - Bright option

## **Top Recommendation for Your Use:**

- ✅ Completely safe for all color vision types
- ✅ High contrast against white/light backgrounds
- ✅ Professional appearance
- ✅ Very different from blues/greens commonly used for controls

```python
# Visual demonstration of alternative color-blind friendly colors
import matplotlib.pyplot as plt
import numpy as np

# Alternative colors that avoid blue, green, purple, red, black, yellow
colors = {
    'Dark Orange': '#FF8C00',
    'Burnt Orange': '#CC5500', 
    'Chocolate Brown': '#8B4513',
    'Sandy Brown': '#F4A460',
    'Dark Turquoise': '#00CED1',
    'Slate Gray': '#708090'
}

# Create a demonstration plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Color swatches
x_pos = np.arange(len(colors))
for i, (name, color) in enumerate(colors.items()):
    ax1.barh(i, 1, color=color, alpha=0.8, edgecolor='white', linewidth=2)
    ax1.text(0.5, i, f'{name}\n{color}', ha='center', va='center', 
             fontweight='bold', color='white', fontsize=10)

ax1.set_yticks(x_pos)
ax1.set_yticklabels(colors.keys())
ax1.set_xlim(0, 1)
ax1.set_title('Color-Blind Friendly Alternatives\n(Avoiding blue, green, purple, red, black, yellow)', 
              fontsize=14, fontweight='bold')
ax1.set_xlabel('Color Samples')

# Example line plot using these colors
x = np.linspace(0, 10, 100)
for i, (name, color) in enumerate(colors.items()):
    y = np.sin(x + i) + i
    ax2.plot(x, y, color=color, linewidth=3, label=name, marker='o', markersize=4, markevery=15)

ax2.set_title('Example Plot with Alternative Colors', fontsize=14, fontweight='bold')
ax2.set_xlabel('X Values')
ax2.set_ylabel('Y Values')
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("🎨 BEST SINGLE RECOMMENDATION: #FF8C00 (Dark Orange)")
print("   ✅ Maximum accessibility")
print("   ✅ Professional appearance") 
print("   ✅ High contrast")
print("   ✅ Completely avoids your excluded colors")
```
